### 4. Rows and columns- Selection 


```python
df[['Employee','Monthly_salary']] #select columns
```

```python
# select rows 
#loc - index name based
df.loc[df.Employee=='Purava'] # for one condition
df.loc[(df.Employee=='Madhav') & (df.Monthly_salary>=100)]
#for mulltiple condition 
```

```python
#iloc - gives index based values 
df.iloc[0]
df.iloc[0:2] #[start:stop:step]
```

```python
 # filter the dataframe

df_age=df[(df['Age']>19) & (df['Monthly_salary']>=1000)] 
```

```python
df_age
```

```python
# where method 
df.where(df['Age']>=19, other="Not eligible")

```

### rows and columns - Operations(add,update,delete)

```python
df
```

```python
# add new column 
df['Team']=['HR','CEO','Tester']
df
```

```python
# other method for column 
df['Bonus']=df['Monthly_salary']*0.2
df
```

```python
len(df)
```

```python
# updating the values 
df.loc[0,'Monthly_salary']=50000
df
```

```python
# other method to update the values 
df.loc[df.Employee=='Madhav','Monthly_Salary']=60000
```

```python
#deleting values
# deleting rows 
df=df.drop(df[df.Employee=='piakchu'].index)
df
```

```python
# deleting columns 
df=df.drop('Monthly_Salary',axis=1)
```
