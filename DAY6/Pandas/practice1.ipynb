Create a Pandas Series using the values:

2.

Create a Series using:

values: 100, 200, 300
index labels: "a", "b", "c"

Access and print the value associated with label "b".

```python
data=pd.Series([100,200,300],index=['a','b','c'])
print(data['b'])

```

3.Create a Series of student marks:

78, 85, 92, 67, 88

Use index labels:

"Aman", "Riya", "Karan", "Neha", "Priya"

Access:

Aman’s marks
Priya’s marks

```python
marks=pd.Series([78,85,92,67,88],index=["Aman", "Riya", "Karan", "Neha", "Priya"])

print(marks["Aman"])
print(marks["Priya"])
```

4.Create a Series:

[5, 10, 15, 20, 25]

Access:

first element
third element
last element

using positional indexing.

```python
numbers=pd.Series([5,10,15,20,25])

print(numbers[0])   
print(numbers[2])   
print(numbers[4])
```

5.Create a Series containing numbers from 1 to 20.

Extract:

first 5 elements
last 5 elements
elements from index 5 to 10

```python
numbers=pd.Series(range(1,21))

#1
print(numbers.iloc[0:5])
print(numbers.iloc[-5:])
print(numbers.iloc[5:11:])
```
