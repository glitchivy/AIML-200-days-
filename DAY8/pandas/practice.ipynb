22.

Create a DataFrame with 10 rows.

Use loc to:

access a single row
access multiple rows
access rows with specific labels

```python
data = pd.DataFrame({
    "Name": ["Aman", "Riya", "Karan", "Neha", "Priya",
             "Rahul", "Sneha", "Arjun", "Pooja", "Vikas"],
    "Age": [21, 22, 20, 23, 24,
            25, 22, 26, 21, 27],
    "Salary": [35000, 42000, 38000, 45000, 50000,
               55000, 40000, 60000, 37000, 65000]
},
index=["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"])

print(data.loc["c"])
print(data.loc[["b", "d", "f"]])
print(data.loc[["a", "e", "j"]])
```
