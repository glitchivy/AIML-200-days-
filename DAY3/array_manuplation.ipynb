# NumPy Array Manipulation Quick Revision

---

# 1. Concatenation

Concatenation means:
```python
joining arrays together
```

Pattern:
```python
np.concatenate((arr1, arr2))
```

Example:
```python
[1,2] + [3,4]
```

Result:
```python
[1,2,3,4]
```

---

# 2. Vertical Stacking

```python
np.vstack((arr1, arr2))
```

Meaning:
```python
stack arrays row-wise
```

Arrays are placed:
```python
top to bottom
```

---

# 3. Horizontal Stacking

```python
np.hstack((arr1, arr2))
```

Meaning:
```python
stack arrays column-wise
```

Arrays are placed:
```python
side by side
```

---

# 4. Splitting Arrays

```python
np.split(arr, parts)
```

Example:
```python
np.split(arr, 2)
```

splits array into:
```python
2 equal parts
```

Important:
```python
array size must divide equally
```

---

# 5. Repeat Elements

```python
np.repeat(arr, times)
```

Repeats:
```python
each element individually
```

Example:
```python
[1,2]
```

Result:
```python
[1,1,2,2]
```

---

# 6. Tile Arrays

```python
np.tile(arr, times)
```

Repeats:
```python
entire array pattern
```

Example:
```python
[1,2]
```

Result:
```python
[1,2,1,2]
```

---

# 7. Unique Values

```python
np.unique(arr)
```

Returns:
```python
sorted unique elements
```

Duplicates removed automatically.

---

# 8. Sorting 1D Arrays

```python
np.sort(arr)
```

Sorts values in ascending order.

---

# 9. Row-wise Sorting

```python
np.sort(arr, axis=1)
```

Each row sorted independently.

---

# 10. Column-wise Sorting

```python
np.sort(arr, axis=0)
```

Each column sorted independently.

---

# 11. Important Axis Reminder

```python
axis=0 -> column direction
axis=1 -> row direction
```

---

# 13. Biggest Beginner Mistakes

## Mistake 1
Confusing:
```python
repeat()
```

with:
```python
tile()
```

### repeat()
duplicates elements.

### tile()
duplicates array pattern.

---

## Mistake 2
Confusing:
```python
concatenate()
```

with:
```python
stack()
```

Stacking usually adds structural dimension organization.

---

## Mistake 3
Forgetting axis meaning during sorting.
