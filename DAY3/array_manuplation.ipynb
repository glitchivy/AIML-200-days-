### Array Manipulation

````python
# Concatenate two arrays.
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print(arr)
````

````python
# Stack arrays vertically.
arr1=np.array([4,5,1])
arr2= np.array([13,24,10])
arr=np.vstack((arr1,arr2))
print(arr)
````

````python
# Stack arrays horizontally.
arr2=([[1],[2],[3]])
arr3=([[5],[6],[7]])

arr=np.hstack((arr2,arr3))
print(arr)
````

````python
# Split an array into equal parts.
arr=np.arange(10)
split_arr=np.split(arr,(2))
print(split_arr)
````

````python
# Repeat elements of an array multiple times.
arr=np.arange(10)
repeat=np.repeat(arr,2)
print(repeat)
````

````python
# Tile an array multiple times.

arr=np.arange(10)
array=np.tile(arr,3)
print(array)
````

````python
# Find unique values in an array.
arr=np.array([1,2,3,44,1,2,3,6,7,8])
array=np.unique(arr)
print(array)
````

````python
# Sort a 1D array.
arr=np.array([2,1,3,1,2,534,32,8923])
arr.sort()
print(arr)
````

````python
# Sort values row-wise in a matrix. 

arr=np.random.randint(1,10,(3,3))
array=np.sort(arr,axis=1)
print(array)
````

````python
# Sort values column-wise in a matrix
arr=np.random.randint(5,10,(3,3))
array=np.sort(arr,axis=0)
print(array)
````

# NumPy Array Manipulation Quick Revision

---

# 1. Concatenation

Concatenation means:
```python
joining arrays together
```

Pattern:
```python
np.concatenate((arr1, arr2))
```

Example:
```python
[1,2] + [3,4]
```

Result:
```python
[1,2,3,4]
```

---

# 2. Vertical Stacking

```python
np.vstack((arr1, arr2))
```

Meaning:
```python
stack arrays row-wise
```

Arrays are placed:
```python
top to bottom
```

---

# 3. Horizontal Stacking

```python
np.hstack((arr1, arr2))
```

Meaning:
```python
stack arrays column-wise
```

Arrays are placed:
```python
side by side
```

---

# 4. Splitting Arrays

```python
np.split(arr, parts)
```

Example:
```python
np.split(arr, 2)
```

splits array into:
```python
2 equal parts
```

Important:
```python
array size must divide equally
```

---

# 5. Repeat Elements

```python
np.repeat(arr, times)
```

Repeats:
```python
each element individually
```

Example:
```python
[1,2]
```

Result:
```python
[1,1,2,2]
```

---

# 6. Tile Arrays

```python
np.tile(arr, times)
```

Repeats:
```python
entire array pattern
```

Example:
```python
[1,2]
```

Result:
```python
[1,2,1,2]
```

---

# 7. Unique Values

```python
np.unique(arr)
```

Returns:
```python
sorted unique elements
```

Duplicates removed automatically.

---

# 8. Sorting 1D Arrays

```python
np.sort(arr)
```

Sorts values in ascending order.

---

# 9. Row-wise Sorting

```python
np.sort(arr, axis=1)
```

Each row sorted independently.

---

# 10. Column-wise Sorting

```python
np.sort(arr, axis=0)
```

Each column sorted independently.

---

# 11. Important Axis Reminder

```python
axis=0 -> column direction
axis=1 -> row direction
```
