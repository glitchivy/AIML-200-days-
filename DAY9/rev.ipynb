### Concatenating DataFrames: `pd.concat`

```python
import pandas as pd


data1 = {'Fruit': ['apple', 'banana'], 'Count': [10, 20]}
df1 = pd.DataFrame(data1)

data2 = {'Fruit': ['cherry', 'date'], 'Count': [15, 25]}
df2 = pd.DataFrame(data2)


df_combined = pd.concat([df1, df2])
print(df_combined)
```
