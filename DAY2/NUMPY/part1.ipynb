```python
# Convert a 3D array into a 1D array.

import numpy as np 

three_dim= np.arange(8).reshape(2,2,2)
three_dim.ndim

arr=three_dim.flatten()
print(arr)
```

```python
# Transpose a matrix.
arr=np.array([[1,2,3,4],[5,6,2,1]])

neew_arr=arr.T
print(neew_arr)
```

```python
# Swap axes in a 3D array.
''' 
Axis 0 → 2 blocks
Axis 1 → 3 rows
Axis 2 → 4 columns
'''
# np.swapaxes(array, axis1, axis2)

arr=np.arange(24).reshape(2,3,4)
print(arr.shape)


new= np.swapaxes(arr,1,2)
print(new.shape)

```

```python
# Add an extra dimension to a 1D array.

arr=np.array([1,2,3,4])
arr.shape
new=arr.reshape(4,1)
new.shape
```

```python
# Remove dimensions of size 1 from an array.

arr=np.array([[1,2,3,4]])
arr.shape
new=arr.squeeze()
new.shape

'''
reshape() when you know target shape
squeeze() when removing singleton dimensions automatically
'''
```
