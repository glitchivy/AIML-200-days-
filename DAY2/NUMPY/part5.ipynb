### Aggregation Functions 


````python
import numpy as np 
````

````python
# Find the sum of all elements in an array.
arr=np.arange(10)
print(np.sum(arr))
````

````python
# Calculate the mean of array values.
arr=np.arange(10)
print(np.mean(arr))
````

````python
# Calculate the median of array values.
arr=np.arange(10)
print(np.median(arr))
````

````python
# Calculate the standard deviation of array values.

arr=np.arange(10)
print(np.std(arr))
````

````python
# Calculate the variance of array values.
arr=np.arange(10)
print(np.var(arr))
````

# NumPy Aggregation Functions Quick Revision

---

# 1. What Are Aggregation Functions?

Aggregation functions:
```python
reduce multiple values into one value
```

Example:
```python
[1,2,3,4]
```

Sum becomes:
```python
10
```

---

# 2. Sum of Elements

```python
np.sum(arr)
```

Adds all elements.

Example:
```python
[1,2,3]
```

Result:
```python
6
```

---

# 3. Mean (Average)

```python
np.mean(arr)
```

Formula:
```python
sum of elements / total elements
```

Example:
```python
[2,4,6]
```

Result:
```python
4
```

---

# 4. Median

```python
np.median(arr)
```

Meaning:
```python
middle value after sorting
```

Example:
```python
[1,3,5]
```

Median:
```python
3
```

---

# 5. Standard Deviation

```python
np.std(arr)
```

Measures:
```python
how spread out data is
```

Small std:
```python
values close together
```

Large std:
```python
values spread apart
```

---

# 6. Variance

```python
np.var(arr)
```

Measures:
```python
spread of data
```

Important:
```python
variance = std²
```

---

# 7. Important Relationship

```python
std = sqrt(variance)
```

Very important statistical relationship.

---

# 8. Axis in Aggregation

Aggregation can happen:
- on entire array
- row-wise
- column-wise

Example:
```python
np.sum(arr, axis=0)
```

Column-wise sum.

Example:
```python
np.sum(arr, axis=1)
```

Row-wise sum.

---

# 9. Important Mental Model

Aggregation means:
```python
many values -> fewer values
```

Sometimes:
```python
many -> one
```

Sometimes:
```python
many -> row summaries
```

---

# 10. Biggest Beginner Mistakes

## Mistake 1
Confusing mean and median.

### Mean
Uses all values.

### Median
Uses middle position.

---

## Mistake 2
Ignoring axis parameter.

Very important in ML tensors.

---

## Mistake 3
Thinking variance and std are unrelated.

They are directly connected:
```python
variance = std²
```
