###  Indexing: Setting Index

```python
import pandas as pd
df = pd.DataFrame({
  "Name": ["Alice", "Bob", "John"],
  "Age": [25, 22, 30],
  "City": ["New York", "Los Angeles", "Chicago"]
})

df.set_index("Name",inplace=True)
print(df)
```

### Indexing: Resetting Index

```python
df.reset_index(inplace=True)
print(df)
```
