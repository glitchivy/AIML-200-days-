```python
import pandas as pd

# DataFrame creation
data = pd.DataFrame({
    'friends': ['Alice', 'Bob', 'Charlie', 'Dave', 'Eve'],
    'scores': [93, 89, 82, 88, 94],
    'age': [20, 21, 20, 19, 21]
})

#putting stats

print("Mean",data['scores'].mean())
print("Mode",data['scores'].mode())
print("Median",data['scores'].median())
print("Standard deviation",data['scores'].std())
print("Varieance",data['scores'].var())
print("Minimum value",data['scores'].min())
print("Maximum values",data['scores'].max())
print('25% Quantile:', data['scores'].quantile(0.25))
```

```python
print(data.describe())
```
