Question 2: Create a Series with Custom Index Labels
Problem Statement

Create a Pandas Series using the following data:

[95, 88, 76]

Use these custom index labels:

["Math", "Science", "English"]

```python
import pandas as pd
data=[95, 88, 76]
series=pd.Series(data,index=["Math", "Science", "English"])
print(series)
```
